```python
def extract_ans(annotated):
    annotated = df
    recipe_ent_dict = {}
    find_cluster_all = {} ##
    for idx in range(len(df['Answer.taskAnswers'])):
        #print("This is recipe", idx, ": ")
        recipe = "recipe_" + str(idx)
        ent_dict = {}
        find_cluster_mention = [] ##
        find_cluster_entity = [] ##
        ans = ast.literal_eval(df['Answer.taskAnswers'][idx])
        #print(ans)
        for i in ans[0]['crowd-entity-annotation']['entities']:
            
            #print("startOffset: ", i['startOffset'])
            #print("endOffset: ", i['endOffset'])
            #print("label: ", i['label'])
            

            startOffset = i['startOffset']+1
            endOffset = i['endOffset']+1

            #print(df['Input.text'][idx][startOffset:endOffset])
            #print()       
            entity = i['label']
            mention =  [df['Input.text'][idx][startOffset:endOffset]]
            #print(entity)
            #print(mention)
            find_cluster_mention.append(mention[0]) ##
            find_cluster_entity.append(entity) ##
            if entity not in ent_dict:
                ent_dict[entity] = mention
            else:
                ent_dict[entity] += mention
            
            recipe_ent_dict[recipe] = ent_dict
            find_cluster_all[recipe] = find_cluster_mention, find_cluster_entity
    #print(find_cluster_all)
    return recipe_ent_dict, find_cluster_all
        
```

```python
clusters_dict = {}
clusters = []
for i in range(len(find_cluster_entity)):
    if find_cluster_entity[i] not in clusters_dict:
        clusters_dict[find_cluster_entity[i]] = []
    clusters_dict[find_cluster_entity[i]].append(find_cluster_wc[i])
    
print(clusters_dict)
    
for entity, not_sorted_np in clusters_dict.items():
    sorted_np = []
    tmp = []
    
    if len(not_sorted_np) == 1:
            sorted_np.append([not_sorted_np[0], not_sorted_np[0]])
            
    for i in range(len(not_sorted_np)-1): #why -1? - because need to count one after another 
        #print(not_sorted_np[i])
        
        if not_sorted_np[i+1] - not_sorted_np[i] == 1: # it means 2 words are neighbors or only one word in the cluster
            if i + 2 == len(not_sorted_np): #end of the list
                if len(tmp) == 0: #nothing in tmp
                    sorted_np.append([not_sorted_np[i], not_sorted_np[i+1]])

                else: #add recoed in tmp and append to sorted list
                    tmp += [not_sorted_np[i], not_sorted_np[i+1]]
                    sorted_np.append(tmp)
                    tmp = []

            else: #not the end of the list
                tmp.append(not_sorted_np[i])
                #print("tmp: ", tmp)
            
            
        if not_sorted_np[i+1] - not_sorted_np[i] != 1: #one word a set or many words a set
            if len(tmp) == 0:
                sorted_np.append([not_sorted_np[i], not_sorted_np[i]])
            else:
                tmp += [not_sorted_np[i]]
                sorted_np.append(tmp)
                tmp = []
                
            if i + 2 == len(not_sorted_np): #the last word
                sorted_np.append([not_sorted_np[i+1], not_sorted_np[i+1]])
        
    #clusters.append(not_sorted_np)
    clusters.append(sorted_np)
    
        
clusters_dict
print()
print('cluster: ',clusters)
#len(clusters)
```

```python
# Here we use "spanbert" as our first model
model_name = "spanbert_base" 

# Determine Max Segment
max_segment = None
for line in open('experiments.conf'):
    if line.startswith(model_name):
        max_segment = True
    elif line.strip().startswith("max_segment_len"):
        if max_segment:
            max_segment = int(line.strip().split()[-1])
            break

tokenizer = tokenization.FullTokenizer(vocab_file="cased_config_vocab/vocab.txt", do_lower_case=False)
subtoken_num = 0
for sent_num, line in enumerate(text):
    raw_tokens = line.split()
    #print(raw_tokens)
    tokens = tokenizer.tokenize(line)
    print(tokens)
    if len(tokens) + len(data['sentences'][-1]) >= max_segment:
        data['sentences'][-1].append("[SEP]")
        data['sentences'].append(["[CLS]"])
        data['speakers'][-1].append("[SPL]")
        data['speakers'].append(["[SPL]"])
        data['sentence_map'].append(sent_num - 1)
        #data['subtoken_map'].append(subtoken_num - 1)
        data['sentence_map'].append(sent_num)
        #data['subtoken_map'].append(subtoken_num)

    ctoken = raw_tokens[0]
    cpos = 0
    for token in tokens:
        data['sentences'][-1].append(token)
        data['speakers'][-1].append("-")
        data['sentence_map'].append(sent_num)
        #data['subtoken_map'].append(subtoken_num)
        
        if token.startswith("##"):
            token = token[2:]
        if len(ctoken) == len(token):
            subtoken_num += 1
            cpos += 1
            if cpos < len(raw_tokens):
                ctoken = raw_tokens[cpos]
        else:
            ctoken = ctoken[len(token):]

data['sentences'][-1].append("[SEP]")
data['speakers'][-1].append("[SPL]")
data['sentence_map'].append(sent_num - 1)
#data['subtoken_map'].append(subtoken_num - 1)
```
